# 5. Division des donnees

```python
data
```

```python
X = data.iloc[ : , : -1 ]
```

```python
y = data.iloc[ : ,  -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=35)
```

```python
print("X-train:", X_train.shape)
print("X-test:", X_test.shape)
print("y-train:", y_train.shape)
print("y-test:", y_test.shape)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler
echelle = StandardScaler()
X_train = echelle.fit_transform(X_train)
X_test = echelle.transform(X_test)
```

```python
X_train
```

# 6. Construction du reseau de neuron superficiel : SNN

```python
# Creation de l'object de la classe Sequential
model_SNN = Sequential()
```

```python
# Ajouter la premiere couche
model_SNN.add(Dense(8, input_shape=(22,), activation='relu'))

# Ajouter de la couche de sortie
model_SNN.add(Dense(1, activation='sigmoid'))

# Compiler le model
model_SNN.compile(optimizer='Adam', 
                  loss='binary_crossentropy', 
                  metrics=['accuracy'])

# affichage de l'architecture du Model
model_SNN.summary()
```

```python
history = model_SNN.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```

```python
#data.head()
```

```python
epoch_accuracy = history.history['accuracy']
epoch_loss = history.history['loss']

epoch_val_accuracy = history.history['val_accuracy']
epoch_val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(0, len(epoch_accuracy)),epoch_accuracy, color='blue', label='Training Accuracy')
plt.plot(range(0, len(epoch_val_accuracy)),epoch_val_accuracy, color='red', label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Visualisation de Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(0, len(epoch_loss)),epoch_loss, color='blue', label='Training Loss')
plt.plot(range(0, len(epoch_val_loss)),epoch_val_loss, color='red', label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Visualisation de Loss')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(0, len(epoch_accuracy)),epoch_accuracy, color='blue', label='Training Accuracy')
plt.plot(range(0, len(epoch_loss)),epoch_loss, color='red', label='Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Value')
plt.title('Visualisation de Accuracy vs Loss')
plt.legend()
plt.show()
```
